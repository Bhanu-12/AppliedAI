**2)Write a program to print twin primes less than 1000. If two consecutive odd numbers are 
both prime then they are known as twin primes**

```python
def prime(n):
    for i in range(2,n):
        if(n%i==0):
            return False
    return True
    
def twins(x):
    for i in range(2,x):
        j = i+2
        if(prime(i) and prime(j)):
            print (i,'and',j)
twins(1000)
```

**3)Write a program to find out the prime factors of a number. Example: prime factors of 56 - 2, 2, 2,7**

```python
import math

def prime(n):
    a= []
    #checking for 2's
    
    while(n%2==0):
        a.append(2)
        n=n/2
    
    #checking for odd primes
    #every composite number has a prime factor less than or equal to its square root
    for i in range(3,int(math.sqrt(n))+1,2):
        while(n%i==0):
            a.append(i)
            n=n/i
    
    #if n is a prime number itself
    
    if(n>2):
        a.append(n)
        
    return a

        
print(prime(56))  
print(prime(17))
print(prime(90))
```


**4)Write a program to implement these formulae of permutations and combinations. 
Number of permutations of n objects taken r at a time: p(n, r) = n! / (n-r)!. Number of combinations of n objects taken r at a time is: c(n, r) = n! / (r!*(n-r)!) = p(n,r) / r!**

```python
def fact(n):
    if(n==1):
        return 1
    elif(n<0):
        print('no factorial')
    else:
        return n * fact(n-1)
def npr(n,r):
    return (fact(n)/fact(n-r))

def ncr(n,r):
    return (npr(n,r)/fact(r))

print(npr(6,3))

print(ncr(9,2))
```

**5)Write a function that converts a decimal number to binary number**

```python
def decimal_to_binary(n):
    a=[]
    while(n>1):
        a.append(n%2)
        n = n//2
    a.append(1)
    return a[::-1]

print(decimal_to_binary(17))
print(decimal_to_binary(56))
        
```

**6)Write a function cubesum() that accepts an integer and returns the sum of the cubes of 
individual digits of that number. Use this function to make functions PrintArmstrong() and isArmstrong() to print Armstrong numbers and to find whether is an Armstrong number**

```python
def cubesum(n):
    sum=0
    while(n!=0):
        sum = sum + ((n%10)**3)
        n = n//10
    return sum
#since the question is explicitly asking to use the fucntion cubesum in other function, 
#only three digit armstrong numbers  are possible

def PrintArmStrong(n):
    for i in range(100,n):
        if(cubesum(i)==i):
            print(i)

PrintArmStrong(1000)

def isArmstrong(n):
    if(cubesum(n)==n):
        print(n,'is an armstrong number')
    else:
        print(n,'is not an armstrong number')
        
isArmstrong(123)
isArmstrong(153)
```

**10)A number is called perfect if the sum of proper divisors of that number is equal to the number. For example 28 is perfect number, since 1+2+4+7+14=28. Write a program to print all the perfect numbers in a given range**

```python
def perfect(n):
    a=[]
    for i in range(1,n):
        if(n%i==0):
            a.append(i)
    return sum(a)

def isperfect(n):
    for i in range(1,n+1):
        if(perfect(i)==i):
            print(i)

isperfect(1000)
```

each is equal to the other number. For example 220 and 284 are amicable numbers.

```python
def Amicable(n):
    a=[]
    for i in range(1,n):
        if(n%i==0):
            a.append(i)
    return a

def isAmicable(x,y):
    if(sum(Amicable(x))==y and sum(Amicable(y))==x):
        
        print(x,'and',y,'are amicable numbers')
    else:
        print(x,'and',y,'are not amicable numbers')



isAmicable(56,284)
isAmicable(220,284)
```

**12)Write a program which can filter odd numbers in a list by using filter function**

```python
a= [1,2,3,4,5,6,7,8,9]

def fil(n):
    if(n%2==0):
        return False
    else:
        return True

b =filter(fil,a)
y = []
for i in b:
    y.append(i)
    
print(y)
```

**14)write a program which can map() and filter() to make a list whose elements are cube of even number in a given list**

```python
a= [1,2,3,4,5,6,7,8,9]

def fil(n):
    if(n%2==0):
        return False
    else:
        return True

def cube(n):
    return n**3

d = map(cube,filter(fil,a))

z= []

for i in d:
    z.append(i)

print(z)
```
